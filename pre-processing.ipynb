```python
import pandas as pd
import numpy as np
import re
from nltk.corpus import stopwords
from nltk import RSLPStemmer
import spacy
```

```python
df = pd.read_excel('BD_analises_funcionais.xlsx')
```

```python
df.head()
```

### Passo 1: Alterar todos caracteres para minúsculo

```python
df_step_1 = df.applymap(str.lower, na_action='ignore')
```

```python
df_step_1.head()
```

### Passo 2: Converte números para sua representação por extenso

```python
numbers = []
for (_,serie) in df_step_1.iterrows():
    if type(serie['antecedente']) == str:
        antecedente = re.findall("\d+[\.\,]?\d*", serie['antecedente'])
        
    if type(serie['comportamento']) == str:
        comportamento = re.findall("\d+[\.\,]?\d*", serie['comportamento'])
        
    if type(serie['consequencia']) == str:
        consequencia = re.findall("\d+[\.\,]?\d*", serie['consequencia'])
    
        
    if type(serie['efeito']) == str:
        efeito = re.findall("\d+[\.\,]?\d*", serie['efeito'])
    
    numbers.extend(antecedente)
    numbers.extend(comportamento)
    numbers.extend(consequencia)
    numbers.extend(efeito)
```

```python
numbers = list(set(numbers))
print(numbers)
```

```python
numbers_map = {
    '3.000': 'três mil',
    '1': 'um',
    '20': 'vinte'
}
```

```python
def replace_number_by_str(x):
    if type(x)!=str:
        return x
    
    result = x
    for i in range(len(numbers)):
        result = re.sub(numbers[i], numbers_map[numbers[i]], result)
    
    return result
```

```python
df_step_2 = df_step_1.applymap(lambda x: replace_number_by_str(x))
```

```python
numbers_2 = []
for (_,serie) in df_step_2.iterrows():
    if type(serie['antecedente']) == str:
        antecedente = re.findall("\d+[\.\,]?\d*", serie['antecedente'])
        
    if type(serie['comportamento']) == str:
        comportamento = re.findall("\d+[\.\,]?\d*", serie['comportamento'])
        
    if type(serie['consequencia']) == str:
        consequencia = re.findall("\d+[\.\,]?\d*", serie['consequencia'])
    
        
    if type(serie['efeito']) == str:
        efeito = re.findall("\d+[\.\,]?\d*", serie['efeito'])
    
    numbers_2.extend(antecedente)
    numbers_2.extend(comportamento)
    numbers_2.extend(consequencia)
    numbers_2.extend(efeito)
```

```python
set(numbers_2)
```

## Passo 3: Remover caracteres que não são letras ou espaços em branco

```python
def reduce_to_letters_and_spaces(x):
    if type(x)!=str:
        return x
    
    return re.sub('[^a-záàâãéèêíïóôõöúçñ ]', ' ', x)
```

```python
df_step_3 = df_step_2.applymap(lambda x: reduce_to_letters_and_spaces(x))
df_step_3['processo'] = df_step_2['processo']
```

```python
df_step_3.head()
```

## Passo 4: Reduzir espaços

```python
def reduce_spaces(x):
    if type(x)!=str:
        return x
    
    partial_result = x.strip()
    
    return " ".join(partial_result.split())
```

```python
df_step_4 = df_step_3.applymap(lambda x: reduce_spaces(x))
```

```python
df_step_4.head()
```

## Passo 5: Remover stop words

```python
def remove_stop_words(x):
    if type(x)!=str:
        return x
    
    stop_words = stopwords.words('portuguese')
    text_words = x.split()
    
    partial_result = [word for word in text_words if word not in stop_words]
    
    return " ".join(partial_result)
```

```python
df_step_5 = df_step_4.applymap(lambda x: remove_stop_words(x))
df_step_5['processo'] = df_step_4['processo']
```

```python
df_step_5.head()
```

## Passo 6.1: Stemming

```python
def stem(x):
    if type(x)!=str:
        return x
    
    stemmer = RSLPStemmer()
    text_words = x.split()
    
    partial_result = [stemmer.stem(word) for word in text_words]
    
    return " ".join(partial_result)
```

```python
df_step_6_1 = df_step_5.applymap(lambda x: stem(x))
df_step_6_1['processo'] = df_step_5['processo']
```

```python
df_step_6_1.head()
```

## Passo 6.2: Lemmatization

```python
def lemmatize(x):
    if type(x)!=str:
        return x
    
    nlp = spacy.load('pt_core_news_sm')
    doc = nlp(x)
    
    partial_result = [token.lemma_ for token in doc]
    
    return " ".join(partial_result)
```

```python
df_step_6_2 = df_step_5.applymap(lambda x: lemmatize(x))
df_step_6_2['processo'] = df_step_5['processo']
```

```python
df_step_6_2.head()
```

```python
df_step_6_1.to_excel('preprocessed_stemming.xlsx')
df_step_6_2.to_excel('preprocessed_lemmatization.xlsx')
```
